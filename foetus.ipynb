```python
import numpy as np 
import pandas as pd 
import os
from pathlib import Path
```

```python
data_root_dir = './data/foetus/Foetus'
database_dir = os.path.join(data_root_dir,'CTG_workshop_database')
n_records = sum([1 for filename in  os.listdir(database_dir) if filename.endswith(".mat") ])
print(f"found {n_records} matlab files in directory {database_dir}")
print("")
print("list of files in data directory:")
print(data_root_dir)
for dirname, _, filenames in os.walk(data_root_dir):
    #print(dirname)
    if "_MACOSX" in dirname:
        continue
    parts = Path(dirname).parts
    level=len(parts)
    print("---"*(level-1), "./"+parts[-1])
    for filename in filenames:
        print("---"*(level),filename)
        


# Any results you write to the current directory are saved as output.
```
